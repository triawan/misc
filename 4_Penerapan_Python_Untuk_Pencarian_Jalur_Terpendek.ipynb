```python
import sys
from heapq import heapify, heappush, heappop

def dijsktra(graph,src,dest):
    inf = sys.maxsize
    data_node = {'A':{'cost':inf,'pred':[]},
    'B':{'cost':inf,'pred':[]},
    'C':{'cost':inf,'pred':[]},
    'D':{'cost':inf,'pred':[]},
    'E':{'cost':inf,'pred':[]},
    'F':{'cost':inf,'pred':[]}
    }
    data_node[src]['cost'] = 0
    kunjungan = []
    temp = src
    for i in range(5):
        if temp not in kunjungan: 
            kunjungan.append(temp)
            min_heap = []
            for j in graph[temp]:
                if j not in kunjungan:
                    cost = data_node[temp]['cost'] + graph[temp][j]
                    if cost < data_node[j]['cost']:
                        data_node[j]['cost'] = cost
                        data_node[j]['pred'] = data_node[temp]['pred'] + list(temp)
                    heappush(min_heap,(data_node[j]['cost'],j))
        heapify(min_heap)
        temp = min_heap[0][1]
    print("Panjang lintasan terpendek: " + str(data_node[dest]['cost']))
    print("Simpul terpendek: " + str(data_node[dest]['pred'] + list(dest)))


if __name__ == "__main__":
    graph = {
        'A':{'B':2,'C':4},
        'B':{'A':2,'C':3,'D':8},
        'C':{'A':4,'B':3,'E':5,'D':2},
        'D':{'B':8,'C':2,'E':11,'F':22},
        'E':{'C':5,'D':11,'F':1},
        'F':{'D':22,'E':1}
    }

    source = 'A'
    destination = 'F'
    dijsktra(graph,source,destination)
```

Penerapan *python* terhadap persoalan *knapsack*

```python
def knapSack(M, wt, p, n):
	if n == 0 or M == 0:
		return 0

	if (wt[n-1] > M):
		return knapSack(M, wt, p, n-1)
	else:
		return max(
			p[n-1] + knapSack(
				M-wt[n-1], wt, p, n-1),
			knapSack(M, wt, p, n-1))

p = [65, 80, 30] # value atau benefit atau keuntungan
wt = [2, 3, 1] # bobot tiap item
M = 5 # kapasitas
n = len(p) # banyaknya item
print(knapSack(M, wt, p, n))
```
